# Exercise 5 : Handle Features

- Use one method between these two methods for the object type columns :-
    1. Convert data type to numeric.
    2. One hot encoding.

```python
def convert(lis): 
    for i in lis:
        df[i] = df[i].astype('category')
    
lis = ['cp', 'ca', 'fbs', 'restecg', 'exang', 'slope', 'present']
convert(lis)
df.dtypes
```

```python
df['cp_new'] = df['cp'].cat.codes 
df['ca_new'] = df['ca'].cat.codes 
df['fbs_new'] = df['fbs'].cat.codes 
df['restecg_new'] = df['restecg'].cat.codes
df['exang_new'] = df['exang'].cat.codes 
df['slope_new'] = df['slope'].cat.codes 
df['present_new'] = df['present'].cat.codes

df
```

```python
from sklearn.preprocessing import OneHotEncoder 

enc = OneHotEncoder()  
  
enc_data = pd.DataFrame(enc.fit_transform(df[['cp', 'ca', 'fbs', 'restecg', 'exang', 'slope', 'present']]).toarray()) 
  
New_df = df.join(enc_data) 

New_df
```

# Exercise 6 : Remove Outliers (If exists)

- Use box plots for some continuous data.

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

axs[0, 0].boxplot(df['trestbps'])
axs[0, 0].set_title('trestbps')

axs[0, 1].boxplot(df['chol'])
axs[0, 1].set_title('chol')

axs[1, 0].boxplot(df['thalach'])
axs[1, 0].set_title('thalach')

axs[1, 1].boxplot(df['oldpeak'])
axs[1, 1].set_title('oldpeak')

plt.show();
```

```python
def limits(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1

    upper = Q3 + 1.5*IQR
    lower = Q1 - 1.5*IQR

    print("Upper Bound:",upper)   
    print("Lower Bound:",lower)
    return upper, lower


print("oldpeak")
upper, lower = limits(df['oldpeak'])
outliers = df[(df['oldpeak'] < lower) | (df['oldpeak'] > upper)]
df = df.drop(outliers.index)
print("--------------------")
print("trestbps")
upper, lower = limits(df['trestbps'])
outliers = df[(df['trestbps'] < lower) | (df['trestbps'] > upper)]
df = df.drop(outliers.index)
print("--------------------")
print("chol")
upper, lower = limits(df['chol'])
outliers = df[(df['chol'] < lower) | (df['chol'] > upper)]
df = df.drop(outliers.index)
print("--------------------")
print("thalach")
upper, lower = limits(df['thalach'])
outliers = df[(df['thalach'] < lower) | (df['thalach'] > upper)]
df = df.drop(outliers.index)
```
